## 기본정제

```python
import pandas as pd
```

```python
df_total21 = pd.DataFrame()
for j in range(2017, 2022,1):
    for i in range(1,13,1):
        file_path = './data/부산광역시소방재난본부/부산광역시금일출동정보({}년{}월).csv'.format(j, i)
        df = pd.read_csv(file_path, encoding = 'euc-kr')
        df_total21 = pd.concat([df_total21,df], axis = 0)
```

```python
df_total21
```

```python
df_total = pd.DataFrame()
for i in range(1,7,1):
    file_path = './data/부산광역시소방재난본부/부산광역시금일출동정보(2022년{}월).csv'.format(i)
    df = pd.read_csv(file_path, encoding = 'utf-8')
    df_total = pd.concat([df_total,df], axis = 0)
```

```python
df_total
```

```python
df_total6 = pd.concat([df_total21,df_total],axis=0)
```

```python
df_total = df_total6.copy()
df_total
```

```python
df_list2 = df_total['재난지점'].str.split(' ')
df_total['재난구'] = df_list2.str.get(1)
df_total['재난동'] = df_list2.str.get(2)
```

```python
df_total
```

```python
df_total['종-구-동'] = df_total['종별'] + ' - ' + df_total['재난구'] + ' - ' + df_total['재난동']
```

```python
df_total
```

```python
total30 = df_total.groupby('종-구-동').count().reset_index()[['종-구-동','종별']].sort_values('종별', ascending=False).reset_index(drop=True).head(30)
```

## 종류별 집계

```python
jong = df_total.groupby('종별').count().reset_index()[['종별','재난규모']].sort_values('재난규모', ascending=False).reset_index(drop=True)
jong
```

```python
jong.columns = ['종별','출동수']
```

```python
import matplotlib.pyplot as plt
```

```python
j = []
for x in range(len(jong)):
    j.append(jong['종별'][x])
j
```

```python
# 폰트지정
plt.rc("font", family = "Malgun Gothic", size=13)
jong.plot(kind='bar')

plt.title("부산광역시 출동종류별 출동수(2017.01 ~ 2022.06)")
plt.ylabel("출동수")

# x축 범위?
a = list(range(0,4,1))
plt.xticks([i for i in range(0,4,1)], j, rotation = 55)
```

## 히트맵

```python
df_total
```

```python
df_p = df_total[['종별','재난구']]
df_p['출동수'] = 1
df_p
```

```python
df_pivot = pd.pivot_table(df_p, index='재난구',columns = '종별', aggfunc= 'count')
```

```python
df_pivot
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_pivot, annot=True, fmt='.0f', cmap='rocket_r')
plt.title("구별 및 출동종류별 히트맵(2017.01 ~ 2022.06)")
plt.show()
```

```python
df_p2 = df_total[['종별','구-동']]
df_p2['출동수'] = 1
df_p2
```

```python
df_pivot2 = pd.pivot_table(df_p2, index='구-동',columns = '종별', aggfunc= 'count')
df_pivot2
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df_pivot2, annot=True,fmt='.0f',  cmap='rocket_r')
plt.title("동별 및 출동종류별 히트맵(2017.01 ~ 2022.06)")
plt.show()
```
