```python
soup
```

```python
cctv_list = soup.find_all('ctltype')
gpsx_list = soup.find_all('gpsx')
gpsy_list = soup.find_all('gpsy')
ym_list = soup.find_all('jbymd')
raddr_list = soup.find_all('rdnwhladdr')
vioaddr_list = soup.find_all('violplcnm')
school_list = soup.find_all('schoolzoneyn')

cctv_list_txt = []
gpsx_list_txt = []
gpsy_list_txt = []
ym_list_txt = [] 
raddr_list_txt = [] 
vioaddr_list_txt = []
school_list_txt = []

for i in range(len(cctv_list)):
    cctv_list_txt.append(cctv_list[i].text)
    gpsx_list_txt.append(gpsx_list[i].text)
    gpsy_list_txt.append(gpsy_list[i].text)
    ym_list_txt.append(ym_list[i].text)
    raddr_list_txt.append(raddr_list[i].text)
    vioaddr_list_txt.append(vioaddr_list[i].text)
    school_list_txt.append(school_list[i].text)
    
new_df = pd.DataFrame({'cctv' : cctv_list_txt, 'x' : gpsx_list_txt,'y':gpsy_list_txt,'year-month':ym_list_txt, \
                       'register': raddr_list_txt, 'vioaddr':vioaddr_list_txt,'schoolyesno' : school_list_txt })
new_df
```
