```python
import pandas as pd
```

```python
file_path = './data/과속방지턱.csv'
df = pd.read_csv(file_path, encoding = 'utf-8-sig')
```

```python
loc = df[['WGS84위도','WGS84경도']]
loc
```

```python
loc.info()
```

```python
loc.columns = ['위도','경도']
loc
```

```python
loc['위경도'] = loc['위도'].astype(str) + loc['경도'].astype(str)
loc
```

```python
gloc = loc.groupby('위경도').head(1).reset_index(drop=True)
gloc
```

```python
gloc[~gloc.duplicated()]
```

```python
gloc = gloc[['위도','경도']]
gloc
```

```python
gloc.to_csv('./data/과속방지턱정제.csv', index = False)
```
