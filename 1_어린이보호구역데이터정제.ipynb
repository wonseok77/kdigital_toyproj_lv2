```python
import pandas as pd
```

```python
gu_list = ['수영구','해운대구','중구','동구','남구','연제구','금정구','부산진구','북구','사하구',\
           '동래구','강서구','서구','사상구','영도구','기장군']
```

```python
df_total = pd.DataFrame()
for i in gu_list:
    file_path = './data/어린이보호구역/부산광역시_{}_어린이보호구역.csv'.format(i)
    df = pd.read_csv(file_path, encoding= 'euc-kr')
    df_total = pd.concat([df_total,df], axis = 0)
```

```python
df_total
```

```python
df_total.info()
```

```python
df_total['위경도'] = df_total['위도'].astype(str) + df_total['경도'].astype(str)
df_total
```

```python
df_total_loc = df_total.groupby('위경도').head(1)
df_total_loc
```

```python
df_total_loc.to_csv('./data/어린이보호구역.csv',encoding='utf-8-sig')
```
