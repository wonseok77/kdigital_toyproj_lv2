```python
import pandas as pd
```

```python
pop = pd.DataFrame({'지역구' : ['중구','서구','동구','영도구','부산진구','동래구',\
                             '남구','북구','해운대구','사하구','금정구','강서구','연제구','수영구','사상구','기장군'], \
                   '유동인구수' : [12.56, 4.36, 7.21, 3.24, 7.09, 5.07, 4.59, 3.93, 5.11, 4.90, 5.63, 11.97, 4.48, 5.18, 6.80, 6.47]})
```

```python
pop
```

```python
df_total21 = pd.DataFrame()
for j in range(2017, 2022,1):
    for i in range(1,13,1):
        file_path = './data/부산광역시소방재난본부/부산광역시금일출동정보({}년{}월).csv'.format(j, i)
        df = pd.read_csv(file_path, encoding = 'euc-kr')
        df_total21 = pd.concat([df_total21,df], axis = 0)
```

```python
df_total = pd.DataFrame()
for i in range(1,7,1):
    file_path = './data/부산광역시소방재난본부/부산광역시금일출동정보(2022년{}월).csv'.format(i)
    df = pd.read_csv(file_path, encoding = 'utf-8')
    df_total = pd.concat([df_total,df], axis = 0)
```

```python
df_total6 = pd.concat([df_total21,df_total],axis=0)
```

```python
df_total = df_total6.copy()
df_total
```

```python
df_list2 = df_total['재난지점'].str.split(' ')
df_total['재난구'] = df_list2.str.get(1)
```

```python
df_total
```

```python
df_gu = df_total.groupby('재난구').count().reset_index()[['재난구','종별']]
```

```python
df_gu.columns = ['지역구','출동수']
```

```python
df_gu
```

```python
mdf = pd.merge(pop,df_gu, left_on='지역구', right_on='지역구', how='left')
mdf
```
