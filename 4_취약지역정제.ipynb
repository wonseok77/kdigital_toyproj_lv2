```python
import pandas as pd
```

```python
file_path = './data/25grid.csv'
df = pd.read_csv(file_path, encoding = 'utf-8-sig')
```

```python
df
```

```python
df.columns
```

```python
df.columns = ['id', 'left', 'top', 'right', 'bottom', 'fid', 'ADM_SECT_C', 'SGG_NM',
       'SGG_OID', 'COL_ADM_SE', 'GID', '도로폭_평균', '도로길이_평균',
       '최소간격', '최대간격', '신호등개수', '어린이보호구역수', '교차로수',
       '경사도','과속방지턱개수','안전센터근처여부','2.5도로폭이하개수']
```

```python
df
```

```python
df['도로길이_평균'] = df['도로길이_평균'].fillna(0)
df['최소간격'] = df['최소간격'].fillna(0)
df['최대간격'] = df['최대간격'].fillna(0)
df['경사도'] = df['경사도'].fillna(0)
df['안전센터근처여부'] = df['안전센터근처여부'].fillna(0)
df['2.5도로폭이하개수'] = df['2.5도로폭이하개수'].fillna(0)
```

```python
df
```

```python
for i in range(len(df)):
    if (df['경사도'][i] == '60-100%'):
        df['경사도'][i] = 100
    elif (df['경사도'][i] == '30-60%'):
        df['경사도'][i] = 60
    elif (df['경사도'][i] == '15-30%'):
        df['경사도'][i] = 30
    elif (df['경사도'][i] == '7-15%'):
        df['경사도'][i] = 15
    elif (df['경사도'][i] == '2-7%'):
        df['경사도'][i] = 7
    elif (df['경사도'][i] == '0-2%'):
        df['경사도'][i] = 2
    elif (df['경사도'][i] == '0'):
         df['경사도'][i] = 0
```

```python
df['경사도'] = df['경사도'].astype(int)
df.info()
```

```python
df['경사도'].value_counts()
```

```python
df = df.drop(['최소간격','최대간격'],axis=1)
```

```python
x_train = df.iloc[:,11:]
grid = df.iloc[:,:11]
```

```python
cor = x_train.corr(method='pearson')
cor
```

```python
# minmax 스케일러
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
```

```python
len(x_train_scaled)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled)
x_train_scaled.columns = ['도로폭_평균', '도로길이_평균', '신호등개수', '어린이보호구역수', '교차로수',
       '경사도','과속방지턱개수','안전센터근처여부','2.5도로폭이하개수']
```

```python
x_train_scaled
```

```python
df_scaled = pd.concat([grid, x_train_scaled], axis = 1)
```

```python
df_scaled.reset_index(drop=True)
```

```python
df_scaled['출동취약지수'] = (2 * df_scaled['신호등개수']) + (1.6 * df_scaled['어린이보호구역수']) + (1.3 * df_scaled['교차로수']) +\
                            (2 * df_scaled['경사도']) + df_scaled['과속방지턱개수'] + (2.5 * df_scaled['2.5도로폭이하개수']) -\
                            (1.8 * df_scaled['도로폭_평균']) - (df_scaled['도로길이_평균'])  -(3*df_scaled['안전센터근처여부'])
df_scaled                        
```

```python
df_scaled['신호현시지수'] = 5*(1 + df_scaled['신호등개수']) + df_scaled['출동취약지수']
```

```python
df_scaled
```

```python
df_scaled.to_csv('./data/신호현시지수.csv', encoding = 'utf-8-sig')
```
